```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
x = 1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29
x
```

```python
mean1 = np.mean(x)
mean1
```

```python
std1 = np.std(x)
std1
```

```python
n = len(x)
n
```

```python
confidence = 0.99
df = n - 1  
t_critical = stats.t.ppf(1 - (1 - confidence) / 2, df)
margin_error_t = t_critical * (std1 / np.sqrt(n))
ci_t = (mean1 - margin_error_t, mean1 + margin_error_t)
pop_std = 0.2 
z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)
margin_error_z = z_critical * (pop_std / np.sqrt(n))
ci_z = (mean1 - margin_error_z, mean1 + margin_error_z)
print("Sample Mean: {:.3f}".format(mean1))
print("Sample Standard Deviation: {:.3f}".format(std1))
print("\n99% Confidence Interval using t-distribution (sample std):")
print("Lower Bound: {:.3f}, Upper Bound: {:.3f}".format(ci_t[0], ci_t[1]))

print("\n99% Confidence Interval using z-distribution (known population std = 0.2):")
print("Lower Bound: {:.3f}, Upper Bound: {:.3f}".format(ci_z[0], ci_z[1]))
```
