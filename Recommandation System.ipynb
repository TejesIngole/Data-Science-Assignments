```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder instance
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit and transform the 'type' column
type_encoded = encoder.fit_transform(df[['type']])

# Create a DataFrame from the encoded data with appropriate column names
type_encoded_df = pd.DataFrame(type_encoded, columns=encoder.get_feature_names_out(['type']))

# Drop the original 'type' column and concatenate the new one-hot encoded columns
df = df.drop('type', axis=1)
df = pd.concat([df, type_encoded_df], axis=1)

display(df.head())
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Split the genres into lists
df['genre'] = df['genre'].str.split(', ')

# Create a MultiLabelBinarizer instance
mlb = MultiLabelBinarizer()

# Fit and transform the 'genre' column
genre_encoded = mlb.fit_transform(df['genre'])

# Create a DataFrame from the encoded data with appropriate column names
genre_encoded_df = pd.DataFrame(genre_encoded, columns=mlb.classes_)

# Drop the original 'genre' column and concatenate the new multi-label encoded columns
df = df.drop('genre', axis=1)
df = pd.concat([df, genre_encoded_df], axis=1)

display(df.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(anime_name, df, similarity_threshold=0.5):

    try:
        anime_index = df[df['name'] == anime_name].index[0]
    except IndexError:
        return "Anime not found in the dataset."

    anime_features = df.drop('name', axis=1).iloc[anime_index]

    cosine_sim = cosine_similarity(anime_features.values.reshape(1, -1), df.drop('name', axis=1))

    similarity_scores = list(enumerate(cosine_sim[0]))

    recommended_anime_indices = [i[0] for i in similarity_scores if i[1] >= similarity_threshold and i[0] != anime_index]

    return df.iloc[recommended_anime_indices][['name']]
```

Now you can use the `recommend_anime` function to get recommendations. For example, to get recommendations for 'Kimi no Na wa.':

```python
recommendations = recommend_anime('Kimi no Na wa.', df)
display(recommendations)
```

```python
def evaluate_recommendations_similarity(anime_name, df, similarity_threshold=0.5):
    try:
        anime_index = df[df['name'] == anime_name].index[0]
    except IndexError:
        print("Anime not found in the dataset.")
        return None

    anime_features = df.drop('name', axis=1).iloc[anime_index]
    cosine_sim = cosine_similarity(anime_features.values.reshape(1, -1), df.drop('name', axis=1))
    similarity_scores = list(enumerate(cosine_sim[0]))

    recommended_anime_scores = [i[1] for i in similarity_scores if i[1] >= similarity_threshold and i[0] != anime_index]

    if not recommended_anime_scores:
        print(f"No recommendations found for '{anime_name}' with similarity threshold {similarity_threshold}.")
        return None

    return sum(recommended_anime_scores) / len(recommended_anime_scores)

anime_to_evaluate = 'Kimi no Na wa.'
average_similarity = evaluate_recommendations_similarity(anime_to_evaluate, df)

if average_similarity is not None:
    print(f"Average similarity of recommendations for '{anime_to_evaluate}': {average_similarity:.4f}")
```

1. Can you explain the difference between user-based and item-based collaborative filtering?
User-Based Collaborative Filtering:
Focuses on finding similar users.
Example: If User A and User B have similar purchase or rating history, then items liked by User B can be recommended to User A.
Challenge: Not scalable for very large datasets (millions of users).
Item-Based Collaborative Filtering:
Focuses on finding similar items.
Example: If many users who buy Item X also buy Item Y, then Item Y is recommended when someone buys Item X.
More scalable and often used in real-world systems (e.g., Amazon).
Key difference: User-based looks at user similarity, item-based looks at item similarity.

2. What is Collaborative Filtering, and how does it work?

Collaborative Filtering (CF) is a recommendation technique that suggests items to a user based on the preferences or behaviors of other users.
It works by:
Collecting user–item interaction data (ratings, purchases, clicks).
Identifying patterns of similarity (between users or items).
Making recommendations based on those similarities.
 Example: If Tejes and Rahul both liked 4 of the same movies, and Rahul liked a 5th movie that Tejes hasn’t seen, the system recommends that 5th movie to Tejes.
