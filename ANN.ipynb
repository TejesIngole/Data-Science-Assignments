```python
# Separate target variable and numerical variables
target = df['letter']
numerical_variables = df.drop('letter', axis=1)

print("Target variable:")
display(target.head())

print("\nNumerical variables:")
display(numerical_variables.head())
```

```python
# Generate box plots for numerical variables
for column in numerical_variables.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=numerical_variables[column])
    plt.title(f'Box Plot of {column}')
    plt.show()
```

```python
# Find outliers using the IQR method
outliers = {}
for column in numerical_variables.columns:
    Q1 = numerical_variables[column].quantile(0.25)
    Q3 = numerical_variables[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    column_outliers = numerical_variables[(numerical_variables[column] < lower_bound) | (numerical_variables[column] > upper_bound)]
    if not column_outliers.empty:
        outliers[column] = column_outliers

# Print the number of outliers found
if outliers:
    print("Number of outliers found in the following columns:")
    for column, outlier_df in outliers.items():
        print(f"Column: {column}, Number of outliers: {len(outlier_df)}")
else:
    print("No significant outliers found in the numerical data.")
```

```python
# Fill outliers with the median of the column
for column, outlier_df in outliers.items():
    median_value = numerical_variables[column].median()
    numerical_variables.loc[outlier_df.index, column] = median_value

print("Outliers have been replaced with the median in the numerical variables DataFrame.")
display(numerical_variables.head())
```

```python
# Find and display outliers using the IQR method (after filling)
outliers_after_fill = {}
for column in numerical_variables.columns:
    Q1 = numerical_variables[column].quantile(0.25)
    Q3 = numerical_variables[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    column_outliers_after_fill = numerical_variables[(numerical_variables[column] < lower_bound) | (numerical_variables[column] > upper_bound)] # Use the numerical_variables df after filling
    if not column_outliers_after_fill.empty:
        outliers_after_fill[column] = column_outliers_after_fill

# Print the number of outliers found after filling
if outliers_after_fill:
    print("Number of outliers found in the following columns after filling:")
    for column, outlier_df in outliers_after_fill.items():
        print(f"Column: {column}, Number of outliers: {len(outlier_df)}")
else:
    print("No significant outliers found in the numerical data after filling.")
```

```python
# Drop the 'ybar' and 'yedgex' columns
numerical_variables = numerical_variables.drop(['ybar', 'yedgex'], axis=1)

print("Dropped 'ybar' and 'yedgex' columns from numerical variables:")
display(numerical_variables.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Apply StandardScaler to the numerical variables
scaler = StandardScaler()
scaled_numerical_variables = scaler.fit_transform(numerical_variables)

# Convert the scaled data back to a DataFrame for better readability (optional)
scaled_numerical_variables_df = pd.DataFrame(scaled_numerical_variables, columns=numerical_variables.columns)

print("Scaled numerical variables:")
display(scaled_numerical_variables_df.head())
```

```python
# Get unique values in the target variable 'letter'
unique_letters = target.unique()

print("Unique values in the target variable 'letter':")
print(unique_letters)
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding to the target variable
label_encoder = LabelEncoder()
encoded_target = label_encoder.fit_transform(target)

print("Label Encoding applied to the target variable.")
display(encoded_target[:5]) # Display the first 5 encoded labels to show the result
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled_numerical_variables, encoded_target, test_size=0.2, random_state=42)

print("Shape of training data (features):", X_train.shape)
print("Shape of testing data (features):", X_test.shape)
print("Shape of training data (target):", y_train.shape)
print("Shape of testing data (target):", y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the baseline ANN model
model = Sequential()

# Input layer and a hidden layer
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],))) # Input layer with 128 neurons and ReLU activation
model.add(Dense(64, activation='relu')) # Another hidden layer with 64 neurons and ReLU activation

# Output layer (assuming multi-class classification for letters)
# The number of units in the output layer should be equal to the number of unique letters
num_classes = len(target.unique())
model.add(Dense(num_classes, activation='softmax')) # Output layer with softmax activation for multi-class classification

# Display the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("Model compiled successfully.")
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Define callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', save_best_only=True)

print("Callbacks defined: EarlyStopping and ModelCheckpoint.")
```

```python
# Train the model with callbacks
history = model.fit(X_train, y_train,
                    epochs=100, # You can adjust the number of epochs
                    batch_size=32, # You can adjust the batch size
                    validation_data=(X_test, y_test),
                    callbacks=[early_stopping, model_checkpoint])

print("Model training complete.")
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
# Define the hyperparameter grid
param_grid = {
    'hidden_layer_sizes': [(128,), (128, 64), (128, 64, 32)], # Number of neurons in each hidden layer
    'activation': ['relu', 'tanh'], # Activation function for hidden layers
    'optimizer': ['adam', 'rmsprop'], # Optimizer
    'batch_size': [32, 64], # Batch size
    'epochs': [50, 100] # Number of epochs
}

print("Hyperparameter grid defined.")
display(param_grid)
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Function to create the Keras model for KerasClassifier
def build_classifier(hidden_layer_sizes, activation='relu', optimizer='adam'):
    model = Sequential()
    # Input layer
    model.add(Dense(hidden_layer_sizes[0], activation=activation, input_shape=(X_train.shape[1],)))
    # Hidden layers
    for layer_size in hidden_layer_sizes[1:]:
        model.add(Dense(layer_size, activation=activation))
    # Output layer
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Create a KerasClassifier
keras_classifier = KerasClassifier(model=build_classifier, verbose=0)

# Add the _estimator_type attribute to make it compatible with newer scikit-learn versions
keras_classifier._estimator_type = "classifier"


# Update the param_grid to use the build_fn parameters
# We will use a smaller grid for demonstration purposes to avoid long execution time
smaller_param_grid = {
    'model__hidden_layer_sizes': [(128,), (128, 64)],
    'model__activation': ['relu'],
    'model__optimizer': ['adam'],
    'batch_size': [32, 64],
    'epochs': [50]
}


# Set up Grid Search
grid_search = GridSearchCV(estimator=keras_classifier,
                           param_grid=smaller_param_grid, # Use the smaller grid
                           scoring='accuracy',
                           cv=3) # You can adjust the number of cross-validation folds

print("Grid Search setup complete.")
```

```python
# Plot training and validation accuracy and loss
plt.figure(figsize=(12, 6))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

print("Plots generated for training and validation metrics.")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
from tensorflow.keras.models import load_model

# Load the best saved model
# Assuming 'best_model.keras' was saved by the ModelCheckpoint callback
try:
    trained_model = load_model('best_model.keras')
except Exception as e:
    print(f"Error loading the trained model: {e}")
    print("Please ensure the model was trained and saved correctly as 'best_model.keras'.")
    # Fallback to the 'model' variable if loading fails, though it might still be the unfitted one
    trained_model = model


# Make predictions on the test set using the loaded/trained model
y_pred_probs = trained_model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

# Compute and print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Compute and print the confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
